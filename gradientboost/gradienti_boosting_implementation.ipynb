```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.DataFrame()
```

```python
df
```

```python
np.random.seed(42)
```

```python
X = np.random.rand(100,1)-0.5
```

```python
X
```

```python
y = 3* X[:,0]**2 + 0.05 * np.random.randn(100)
```

```python
df["X"]= X.reshape(100)
```

```python
df["y"]= y
```

```python
df.head()
```

```python
plt.scatter(df["X"],df['y'])
plt.title("X vs y")
```

```python
df['pred1[base_model]']=df["y"].mean()
```

```python
df
```

```python
df ['res1'] =df['y']-df['pred1[base_model]']
```

```python
df
```

```python
# res1 == taerget column
```

```python
plt.scatter(df["X"],df['y'])
plt.title("X vs y")
plt.plot(df["X"],df['pred1[base_model]'],color='green')
```
