```python
# kernel trick
'''Kernel trick is a technique that allows you to apply linear models to nonlinear 
problems by transforming the data into a higher-dimensional feature space. It is often used in machine learning 
for predictive modeling, such as classification, regression, clustering, and dimensionality reduction.'''

```

```python
# linear seperable data ==> soft classifier 
# non linear seperable data ==> Kernel trick ==> mathematical formula => transofmration (lower to higher dimentiion)
```

```python
# Transformation technique 
    # 1. polynomial
    # 2. sigmoid
    # 3. RBF
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from matplotlib.colors import ListedColormap 
```

```python
from sklearn.datasets import make_circles
```

```python
X, y= make_circles(100,factor=.1,noise=.1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s =50, cmap='bwr')
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.20)
```

```python
X_train
```

```python
classsifier=SVC(kernel='linear')
classsifier.fit(X_train,y_train)
y_pred = classsifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
y_test
```

```python
y_pred
```

```python
classsifier2=SVC(kernel='rbf')
classsifier2.fit(X_train,y_train)
y_pred2 = classsifier2.predict(X_test)
```

```python
accuracy_score(y_test, y_pred2)
```

```python
classsifier3=SVC(kernel='poly')
classsifier3.fit(X_train,y_train)
y_pred3 = classsifier3.predict(X_test)
accuracy_score(y_test, y_pred3)
```

```python
classsifier4=SVC(kernel='poly', degree=2)
classsifier4.fit(X_train,y_train)
y_pred4 = classsifier4.predict(X_test)
accuracy_score(y_test, y_pred4)
```
