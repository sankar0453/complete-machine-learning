```python
# kernel trick
'''Kernel trick is a technique that allows you to apply linear models to nonlinear 
problems by transforming the data into a higher-dimensional feature space. It is often used in machine learning 
for predictive modeling, such as classification, regression, clustering, and dimensionality reduction.'''

```

```python
# linear seperable data ==> soft classifier 
# non linear seperable data ==> Kernel trick ==> mathematical formula => transofmration (lower to higher dimentiion)
```

```python
# Transformation technique 
    # 1. polynomial
    # 2. sigmoid
    # 3. RBF
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from matplotlib.colors import ListedColormap 
```

```python
from sklearn.datasets import make_circles
```

```python
X, y= make_circles(100,factor=.1,noise=.1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s =50, cmap='bwr')
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.20)
```

```python
X_train
```

```python
classsifier=SVC(kernel='linear')
classsifier.fit(X_train,y_train)
y_pred = classsifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
y_test
```

```python
y_pred
```

```python
classsifier2=SVC(kernel='rbf')
classsifier2.fit(X_train,y_train)
y_pred2 = classsifier2.predict(X_test)
# https://en.wikipedia.org/wiki/Radial_basis_function
```

```python
accuracy_score(y_test, y_pred2)
```

```python
classsifier3=SVC(kernel='poly')
classsifier3.fit(X_train,y_train)
y_pred3 = classsifier3.predict(X_test)
accuracy_score(y_test, y_pred3)
```

```python
classsifier4=SVC(kernel='poly', degree=2)
classsifier4.fit(X_train,y_train)
y_pred4 = classsifier4.predict(X_test)
accuracy_score(y_test, y_pred4)
```

```python
# Function to plot decision boundaries and return the Matplotlib figure
def plot_decision_boundary(clf, X, y, plot_title="Decision Boundary"):
    h = .02  # step size in the mesh

    # Plot the decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k', marker='o', s=80)
    plt.title(plot_title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # Create a Matplotlib figure and return it
    fig = plt.gcf()
    return fig
```

```python
plot_decision_boundary(classsifier,X,y)
```

```python
plot_decision_boundary(classsifier2,X,y)
```

```python
plot_decision_boundary(classsifier3,X,y)
```

```python
plot_decision_boundary(classsifier4,X,y)
```

```python
def plot_3d_plot(X,y):
    r = np.exp(-(X**2).sum(1))
    ax= plt.subplot(projection='3d')
    ax.scatter3D(X[:,0],X[:,1],r,c=y,s=100,cmap='bwr')
    ax.set_xlabel("X1")
    ax.set_ylabel("X2")
    ax.set_zlabel("y")
    return ax
    
```

```python
plot_3d_plot(X,y)
```
